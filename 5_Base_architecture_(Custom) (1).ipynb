IMPORT LIBRARIES

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout,MaxPool2D
```

```python
# generators
train_ds = keras.utils.image_dataset_from_directory(
    directory = r"D:\NMIMS\Sem III\SEM III PROJECT\140k real and fake dataset\real_vs_fake\real-vs-fake\train",
    labels='inferred',
    label_mode = 'int',
    batch_size=200,
    image_size=(256,256)
)
test_ds = keras.utils.image_dataset_from_directory(
    directory = r"D:\NMIMS\Sem III\SEM III PROJECT\140k real and fake dataset\real_vs_fake\real-vs-fake\test",
    labels = 'inferred',
    label_mode = 'int',
    batch_size = 200,
    image_size = (256,256)
)
```

NORMALIZE IMAGES

```python
# Normalize
def process(image,label):
    image = tf.cast(image/255. ,tf.float32)
    return image,label

train_ds = train_ds.map(process)
test_ds = test_ds.map(process)
```

# model 1


```python
# model with batch normalization and drop out to reduce overfitting of the model 

model = Sequential()

model.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(256,256,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

MODEL COMPILE

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(train_ds,epochs=5,validation_data=test_ds)
```
